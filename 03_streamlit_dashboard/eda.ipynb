# Exploring olympic games dataset


Dataset found here
- [Olympics 124 years Dataset(till 2020)](https://www.kaggle.com/datasets/nitishsharma01/olympics-124-years-datasettill-2020)

```python
import pandas as pd 

df_summer = pd.read_csv("../data/athletes_summer_games.csv", index_col=0)
df_summer
```

```python
df_swe = df_summer.query("NOC == 'SWE'")

df_swe_medals = df_swe[df_swe["Medal"].notna()]
df_swe_medals
```

```python
df_summer[df_summer["Medal"].notna()].groupby("NOC")["Medal"].count().sort_values(
    ascending=False
)
```

```python
total_medals_swe = len(df_swe_medals)
total_medals_swe
```

```python
df_swe_medals.groupby("Name")["Medal"].count().sort_values(ascending=False)
```
