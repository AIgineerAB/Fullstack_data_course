## Matplotlib

```python
import matplotlib.pyplot as plt

f = lambda x: 3 * x - 2

def g(x):
    return x**2 - 2

x = tuple(range(-5,5))

y1 = [f(i) for i in x]
y2 = [g(i) for i in x]

plt.plot(x, y1)
plt.plot(x, y2)
plt.title("plotting 2 fcns")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["$f(x)=3x-2$", "$g(x) = x^2 - 2$"])
```

```python
f = lambda x: x**2 + 1
g = lambda x: np.sin(x) + 5

# OOP approach 
fig, ax = plt.subplots()
ax.plot(x, f(x), label = "$f(x) = x^2 + 1$")
ax.plot(x, g(x), label = "$g(x) = \sin x + 5$")
ax.set(title = "Two graphs", xlabel="x", ylabel="y")
ax.legend()
```

## File handling

```python
ml_text_path = "ml_text_raw.txt"

with open(ml_text_path, "r") as file:
    raw_text = file.read()

raw_text
```

```python
print(raw_text)
```

```python
import re

text_fixed_space = re.sub(r"\s+", " ", raw_text)
text_list = text_fixed_space.split(". ")
text_list
```

```python
text_list_cleaned = [sentence.capitalize() for sentence in text_list]
text_list_cleaned
```

```python
cleaned_text = ".\n".join(text_list_cleaned)
print(cleaned_text)
```

```python
with open("ml_text_cleaned.txt", "w") as file:
    file.write(cleaned_text)
```
