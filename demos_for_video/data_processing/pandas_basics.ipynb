```python
try:
    df.astype({"rating": "float"})
except ValueError as err:
    print(err)
```

```python
df[pd.to_numeric(df["rating"], errors="coerce").isna()]
```

```python
df = df[pd.to_numeric(df["rating"], errors="coerce").notna()]
df.shape
```

```python
df.query("product_id == 'B08L12N5H1'")
```

```python
df.head()
```

```python
df.query("product_id == 'B098NS6PVG'")
```

```python
try:
    df.astype({"rating": float, "rating_count": int})
except ValueError as err:
    print(err)
```

```python
df.loc[:,"rating_count"] = df["rating_count"].str.replace(",", "")
df.head(2)
```

```python
try:
    df.astype({"rating": float, "rating_count": int})
except ValueError as err:
    print(err)
```

```python
df = df[df["rating_count"].notna()]
df.info()
```

```python
try:
    df = df.astype({"rating": float, "rating_count": int})
except ValueError as err:
    print(err)
```

```python
df.info()
```

```python
top_10_most_rated = df.sort_values(by="rating_count", ascending=False).iloc[:10]
top_10_most_rated[["product_name", "rating", "rating_count", "actual_price"]]
```

```python
top_10_most_rated["review_title"].iloc[0].split(",")
```
